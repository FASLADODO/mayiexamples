# Feature Engineering: the Heart of Data Science

我们希望回答诸如“他在过去三个月或过去三次交易中花了多少钱？”或“在购买的所有产品中，杂货的百分比是多少？ 衣服？“等

我们希望有更多的信息来描述群体事件随时间的演变：频率（事件的发生）和新近度（事件之间的间隔）。 所以我们想知道“客户在过去6个月内购买了多少次产品？”或“他最后一次购买产品的时间是什么时候？”



我们的软件包生成多个功能系列，可以生成数百个功能。 除了上面例子中的两个家庭，还有：

* 信息功能：主题状态统计。 例如，在过去n天/周/月中花费的平均金额。

* 分布函数：分类列中不同值的百分比。 例如，星期日发生的事件百分比。

EventsAggregator的另一个重要方面是时间聚合，它存在于两个层面：

按组进行一次性聚合：对于每个组，我们聚合所有事件以便具有该组的一般视图。 流失预测是一个很好的例子，因为我们想知道客户是否会根据他们的历史活动停止使用该产品。

按事件顺序聚合：对于每个组，我们按事件顺序聚合要素。 一个示例用例是欺诈检测，其中我们需要通过事务分析用户的行为事务。


继续推进特征工程

特征工程是数据科学和机器学习中最重要的任务之一。 在现实生活中，它通常有两个目的：提高预测模型的性能，并创建表达指标，使数据科学家/分析师能够更深入地了解手头的问题。

大多数自动特征工程方法往往只关注前一点。 功能通常以强力或不可解释的方式生成（如在深度学习中）。 然而，在许多现实生活中的用例中，利益相关者需要更多; 他们需要可操作的见解，他们需要功能/ KPI /指标，以帮助他们了解接下来应该采取的行动。

我们相信我们的方法可以回答上述两个问题。 自动特征工程不仅仅是为最后0.1％的指标提升优化模型，而是更多关于能够通过更好的功能来讲述更好的故事。
