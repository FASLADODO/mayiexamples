## 模型函数
模型函数（即 model_fn）会实现机器学习算法。

**采用预创建的 Estimator 和自定义 Estimator 的唯一区别是：

* 如果采用预创建的 Estimator，则有人已为您编写了模型函数。
* 如果采用自定义 Estimator，则您必须自行编写模型函数。
